This file is for plotting simulation data.
The simulation files need to be run before they can be plotted.
The headers specify which simulations are being plotted.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Times New Roman"],
    "font.size": 10,
    "axes.titlesize": 10,
    "axes.spines.right": False,
    "axes.spines.top": False,
    "lines.linewidth": 1.75,
    "axes.formatter.limits": [-3,3],
    "axes.formatter.use_mathtext":True
})

col_width = 6
```

# Linear Constraints Large $b$  (Figure 2a)

```python
gnop_short = np.load('short_gnop_reg.npy')
roful_short = np.load('short_roful_reg.npy')
gnop_viol = np.load('short_gnop_viol.npy')
roful_viol = np.load('short_roful_viol.npy')

num_trials = np.shape(roful_short)[0]
T = np.shape(roful_short)[1]
tsteps = np.arange(T) + 1

gnop_norm = np.cumsum(gnop_short,1)/np.sqrt(tsteps)
gnop_mean = np.mean(gnop_norm,0)
gnop_std = np.std(gnop_norm,0)

roful_norm = np.cumsum(roful_short,1)/np.sqrt(tsteps)
roful_mean = np.mean(roful_norm,0)
roful_std = np.std(roful_norm,0)

fig, ax = plt.subplots(figsize=(col_width*0.55,2.3),dpi=300)
ax.plot(tsteps, gnop_mean, label='GenOP')
ax.fill_between(tsteps, gnop_mean - gnop_std, gnop_mean + gnop_std,alpha =0.3)
ax.plot(tsteps, roful_mean, label='ROFUL')
ax.fill_between(tsteps, roful_mean - roful_std, roful_mean + roful_std,alpha =0.3)
ax.legend()
ax.set_ylim(bottom=0)
ax.set_xlabel('Round $t$')
ax.set_ylabel('Sqrt-normalized Regret $R_t/\\sqrt{t}$')
fig.tight_layout()
fig.savefig('slb_short.png')

print('ROFUL Total Violation',np.sum(roful_viol))
print('GenOP Total Violation',np.sum(gnop_viol))
```
